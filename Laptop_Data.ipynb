```python
price = df['Price (Inr)']
plt.figure(figsize=(6,2))
sns.histplot(price,kde=True)
```

```python
top_publishers = df.groupby('Company')['Price (Inr)'].sum().nlargest(8)
plt.figure(figsize=(6,3))
top_publishers.plot(kind='pie',title='Manufacturars Selling price ration')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x = df['OpSys'], y = df['Price (Inr)'])
plt.title('Operating System & Price (Inr)')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Weight (kg)', y='Price (Inr)', data=df)
plt.title('Weight vs Laptop Price')
plt.xlabel('Weight (kg)')
plt.ylabel('Price (Inr)')
plt.show()
```

```python
x = df.corr(numeric_only=True)
plt.figure(figsize=(5, 2))
sns.heatmap(x, annot=True, cmap='crest')

plt.show()
```

```python
company_avg_price = df.groupby('Company')['Price (Inr)'].mean()
plt.figure(figsize=(8, 3))
plt.bar(company_avg_price.index, company_avg_price.values, color='blue')
plt.title('Average Laptop Price by Company')
plt.xlabel('Company')
plt.ylabel('Average Price (Inr)')
plt.xticks(rotation=45)
plt.show()
```
